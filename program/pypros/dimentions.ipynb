```python
import pandas as pd
import openpyxl
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
filename="/Users/kazuki/Documents/Study/data/ryukyu/typeFreq.csv"
```

```python
df=pd.read_csv(filename,header=0, index_col=0)
```

```python
print(df)
```

```python
datasets=df.values
```

```python
pred=KMeans(n_clusters=5, init='k-means++', 
              n_init=10, max_iter=300,verbose=1).fit_predict(datasets)
```

```python
print(pred)
```

```python
locationfile="/Users/kazuki/Documents/Study/data/ryukyu/locations.xlsx"
ldf=pd.read_excel(locationfile,header=0,index_col=0)
```

```python
for i,clnum in enumerate(pred):
    print("{}  {}".format(clnum,ldf.iat[i,1]))
```

### クラスター番号の保存

```python
ldf["cluster_id"]=pred
```

```python
cl0=ldf.query("cluster_id == 0")
cl1=ldf.query("cluster_id == 1")
cl2=ldf.query("cluster_id == 2")
cl3=ldf.query("cluster_id == 3")
cl4=ldf.query("cluster_id == 4")
```

```python
cl0=cl0["現市町村名"]
cl1=cl1["現市町村名"]
cl2=cl2["現市町村名"]
cl3=cl3["現市町村名"]
cl4=cl4["現市町村名"]
```

```python
cl0=cl0.values
cl1=cl1.values
cl2=cl2.values
cl3=cl3.values
cl4=cl4.values
```

```python
print(len(cl0))
print(len(cl1))
print(len(cl2))
print(len(cl3))
print(len(cl4))
```

```python
a=["0"]*(len(cl2)-len(cl0))
cl0=np.append(cl0,a)
a=["0"]*(len(cl2)-len(cl1))
cl1=np.append(cl1,a)
a=["0"]*(len(cl2)-len(cl3))
cl3=np.append(cl3,a)
a=["0"]*(len(cl2)-len(cl4))
cl4=np.append(cl4,a)
```

```python
cl0=pd.DataFrame(cl0,columns=["0"])
cl1=pd.DataFrame(cl1,columns=["1"])
cl2=pd.DataFrame(cl2,columns=["2"])
cl3=pd.DataFrame(cl3,columns=["3"])
cl4=pd.DataFrame(cl4,columns=["4"])
```

```python
clnums=pd.concat([cl0,cl1,cl2,cl3,cl4],axis=1)
```

```python
wfile="/Users/kazuki/Documents/Study/data/ryukyu/kclus.xlsx"
with pd.ExcelWriter(wfile) as writer:
        clnums.to_excel(writer,sheet_name="クラスター")
```

# 正規化してからプロット

```python
sou=np.sum(datasets,axis=1)
sou2=sou[:, np.newaxis]
print(sou2.shape)
```

```python
seidata=datasets/sou2
```

```python
pred=KMeans(n_clusters=5, init='k-means++', 
              n_init=10, max_iter=300,verbose=1).fit_predict(seidata)
```

```python
ldf["cluster_id"]=pred
cl=[]
for i in range(5):
    aaa="cluster_id == "+str(i)
    cl.append(ldf.query(aaa))
```

```python
mlength=[]
for i in range(5):
    cl[i]=cl[i]["現市町村名"]
    cl[i]=cl[i].values
    mlength.append(len(cl[i]))
    print(mlength[i])
maxer=max(mlength)
print(maxer)
```

```python
for i in range(5):
    a=["0"]*(maxer-len(cl[i]))
    print(len(a))
    cl[i]=np.append(cl[i],a)
    cl[i]=pd.DataFrame(cl[i],columns=[str(i)])
clnums=pd.concat(cl,axis=1)
wfile="/Users/kazuki/Documents/Study/data/ryukyu/kclus2.xlsx"
with pd.ExcelWriter(wfile) as writer:
        clnums.to_excel(writer,sheet_name="クラスター")
```
