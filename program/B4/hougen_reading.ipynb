```python
def ngrams(text, n):
    text_list = list(text)
    return [text_list[i:i + n] for i, char in enumerate(text_list)]
```

```python
print(ngrams(list(df.iloc[1,:]),2))
```

```python
testdic={i:{} for i in range(96)}
```

```python
print(testdic)
```

```python
for i in range(96):
    text=list(df.iloc[i,:])
    data=ngrams(text,2)
    for j,d in enumerate(data):
        print(d)
        if (-9) in d:
            break
        elif (-1) in d:
            d[d.index(-1)]="."
        d="-".join(d)
        if d not in testdic[i]:
            testdic[i][d]=1
        else:
            testdic[i][d]+=1
```
