```python
import pandas as pd
import openpyxl
import numpy as np
#from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import scipy.spatial.distance as dis
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as hierarchy
```

```python
basepath ="/Users/kazuki/Documents/Study/data/ryukyu/{}"
basepath2="/Users/kazuki/Downloads/{}"
rftri    = basepath2.fotmat("想定形.xlsx")
rflocate = basepath.format("parameter/locations.xlsx")
rftango  = basepath.format("parameter/sheetlist.xlsx")
```

```python
triDf    = pd.read_excel(rftri   , sheet_name=0, header=0, index_col=0)
locateDf = pd.read_excel(rflocate, sheet_name=0, header=0, index_col=0)
tangoDf  = pd.read_excel(rftango , sheet_name=0, header=0, index_col=0)
```

```python
shNs    = list(triDf.keys())
locates = list(locateDf.iloc[1:,2])
words   = list(tangoDf.iloc[:,0])
```

```python
print(shNs)
```

```python
#類似度を求める
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
allhindovector=[]
for si,shN in enumerate(shNs):
    a=triDf[shN].values
    for sj,(locate,shN2) in enumerate(zip(locates,shNs)):
        if si == sj:
            continue
        b=triDf[shN2].values
        allhindovector.append([locate,cos_sim(a,b)])
```

```python
clusterdata=triDf.values
clusterdata=clusterdata.T
```

```python
#図の生成
figtitle='ウォード法の樹形図'
himethod='ward'
savefile="/Users/kazuki/Documents/Study/picture/wardCluster.png"
z = hierarchy.linkage(clusterdata, method=himethod)
plt.figure(figsize=(60,30))
plt.rcParams['font.family'] = 'IPAexGothic'
plt.rcParams['font.size'] = 50
plt.title(figtitle, fontsize=60)
plt.xlabel('各地点', fontsize=55)
plt.ylabel('クラスタ間の距離', fontsize=55)
# hierarchy.set_link_color_palette(['purple', 'lawngreen', 'green', 'blue', 'orange', 'red']) # ６クラスタまでの色を指定
hierarchy.dendrogram(z, leaf_font_size=35., # 横軸の文字の大きさを指定
                        color_threshold=0., # ユークリッド平方距離が７以上を同色で表示
                        above_threshold_color='black',
                        labels=tuple(locates)) # ユークリッド平方距離が７以上を黒色に指定
plt.tight_layout()
plt.savefig(savefile)
plt.show()
plt.close()
```
