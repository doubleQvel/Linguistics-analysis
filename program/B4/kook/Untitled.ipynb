```python
basepath ="/Users/kazuki/Documents/Study/data/ryukyu/{}"
basepath2="/Users/kazuki/Downloads/{}"
rflocate = basepath.format("parameter/locations.xlsx")
rfword   = basepath.format("parameter/sheetlist.xlsx")
```

```python
locateDf = pd.read_excel(rflocate, sheet_name=0, header=0, index_col=0)
wordDf  = pd.read_excel(rfword , sheet_name=0, header=0, index_col=0)
```

```python
symbols=list(locateDf.iloc[:,0])
locates = list(locateDf.iloc[:,2])
hyouzi = list(locateDf.iloc[:,1])
words   = list(wordDf.iloc[:,0])
```

```python
phoneme=["all","cvc","vcv"]
basyo=symbols[10]
for j in range(3):
    rftri=basepath.format("gramTri/pattern/{}/{}.xlsx")
    triDf=pd.read_excel(rftri.format(phoneme[j],basyo), sheet_name=0, header=0, index_col=0)
    counter=[0 for locate in locates]
    for li,locate in enumerate(locates):
        for i,pat in enumerate(list(triDf.index)):
            moto,hikaku=pat.split("->")
            if moto == hikaku:
                counter[li]+=triDf.iat[i,li]
    for li,locate in enumerate(locates):
        counter[li]=counter[li]/triDf[locate].sum()
        
    newdata=np.array(counter)
    x=locates
    y=newdata
    # plt.rcParams['font.family'] = 'IPAPGothic' #全体のフォントを設定
    plt.rcParams['font.family'] = 'IPAexGothic'
    plt.figure(figsize=(25,10))
    plt.rcParams["font.size"] = 18
    plt.ylim(0.0,1.0)
    
    plt.bar(x,y,align="center",label="無変化")
#     plt.xticks(x,locates)
    plt.xticks(x,locates,rotation='vertical')
    
    plt.legend()
    plt.tight_layout()
    plt.show()
    pictname=basepath2.format("{}muhenka{}.png".format(basyo,phoneme[j]))
    plt.savefig(pictname)
    plt.close()
```
