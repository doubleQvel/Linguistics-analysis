```python
locatedf=pd.read_csv(rflocate2,header=0,index_col=0)
```

```python
locates=list(locatedf.loc["24":"66","0"])
locates1=list(locatedf.loc["24":"32","0"])
locates2=list(locatedf.loc["33":"66","0"])
```

```python
uniAdf2=uniAdf.loc[:,locates[0]:locates[-1]]
```

```python
stA=uniAdf2.values
stA1=uniAdf.loc[:,locates1[0]:locates1[-1]].values
stA2=uniAdf.loc[:,locates2[0]:locates2[-1]].values
```

```python
stA=stA.astype(np.int32)
stA1=stA1.astype(np.int32)
stA2=stA2.astype(np.int32)
```

```python
stAmean  =np.average(stA  , axis=1)
stA1mean=np.average(stA1, axis=1)
stA2mean=np.average(stA2, axis=1)
```

```python
stAvar  =np.var(stA  , axis=1)
stA1var=np.var(stA1, axis=1)
stA2var=np.var(stA2, axis=1)
```

```python
stTest=np.array([stAmean,stA1mean,stA2mean])
```
