```python
import math
import pandas as pd
import openpyxl
import numpy as np
```

```python
rf="/Users/kazuki/Downloads/patternlist.xlsx"
```

```python
df = pd.read_excel(rf, sheet_name=0, header=0, index_col=0)
```

```python
patterns=[0 for i in list(df.index)]
```

```python
ind=list(df.iloc[:,0])
print(ind)
```

```python
for i,pattern in enumerate(ind):
    moto,hikaku=pattern.split("->")
    patterns[i]=[moto,hikaku]
```

```python
print(patterns)
```

```python
with pd.ExcelWriter(rf, engine='openpyxl') as writer:
    wdf=pd.DataFrame(patterns,index=list(df.index), columns=["moto","hikaku"])
    wdf.to_excel(writer,sheet_name="all") #[]
```
