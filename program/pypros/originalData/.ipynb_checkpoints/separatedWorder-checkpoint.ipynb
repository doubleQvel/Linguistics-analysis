# 改良バージョン(nグラムから)

```python
def ngrams(text, n):
    #text:子音と母音の塊
    #[i:i+n]スライス の前にtext最初に<s>を追加
    #list.index(".")で最初の位置を確認　その位置まではスライスを行う if文で？
    text_list = ["<s>"]+list(text)
    if "." not in text_list:
        text_list.append(".")
    saku=text_list.index(".")
    return [text_list[i:i + n] for i, char in enumerate(text_list) if saku+1 >= i+n]
```

```python
#シート内の行数取得
sheetNames=list(df.keys())
raws=len(df[sheetNames[0]])
testdic={}
for i in range(raws):
    text=list(map(str,df[sheetNames[0]].iloc[i,2:]))
    data=ngrams(text,2)
    for j,d in enumerate(data):
        print(d)
        if "-9" in d:
            break
        d="-".join(d)
        if d not in testdic:
            testdic[d]=1
        else:
            testdic[d]+=1
```

```python
class Framer:
    def __init__(self,file):
        self.df=pd.read_excel(file,sheet_name=None)
        self.sheetNames=list(self.df.keys())
        self.tokendict={}
        self.tokenum=0
    def tokener(self,shnum,ng):
        sheet=self.df[self.sheetNames[shnum]]
        raws=len(sheet)
        self.tokendict={}
        for i in range(raws):
            text=list(map(str,sheet.iloc[i,2:]))
            data=self.ngramize(text,ng)
            for j,d in enumerate(data):
                if "-9" in d:
                    break
                elif d.count("-1") >= 2:
                    continue
                d="-".join(d)
                if d not in self.tokendict:
                    self.tokendict[d]=1
                else:
                    self.tokendict[d]+=1
        self.tokenum=len(self.tokendict.keys())
    def ngramize(self,text, n):
        text_list = ["<s>"]+list(text)
        if "-1" in text_list: #-1は無くして圧縮を行う
            text_list=[i for i in text_list if i != "-1"]
        if "." not in text_list:
            text_list.append(".")
        saku=text_list.index(".")
        return [text_list[i:i + n] for i, char in enumerate(text_list) if saku+1 >= i+n]
```
