# やること
1. トライグラムのパターン形式を全て調音方式へと変換
2. シートごとに取り出し、重複して入ればスルー、新規であれば変換
3. 変換し切ったらエクセルに保存
4. できれば地点名だけのシートを作る
5. Rでエクセル読み出し、その後、実行できる形に調整したのち、クラスタリングを行う。
6. クラスタリングを行なった結果より特定のグループを選び、どの地点のものがそこに含まれるパターンを多く使用しているか検討する

```python
filename1="/Users/kazuki/Documents/Study/data/ryukyu/triFreq2.xlsx"
filename2="/Users/kazuki/Documents/Study/data/ryukyu/rule2.xlsx"
wfilename="/Users/kazuki/Documents/Study/data/ryukyu/articulation.xlsx"
```

```python
import pandas as pd
import openpyxl
import numpy as np
```

```python
dft=pd.read_excel(filename1,sheet_name=None,header=0, index_col=0)
# print(dft)
```

```python
dfr1=pd.read_excel(filename2,sheet_name="母音入力数値一覧",header=0, index_col=0)
dfr2=pd.read_excel(filename2,sheet_name="子音入力数値一覧",header=0, index_col=0)
print(dfr1)
print(dfr2)
```

```python
boin=list(dfr1.index)
shiin=list(dfr2.index)
```

```python
pattern=[]
goku=[]
#書くシートから新規パターンを変換
for i,shn in enumerate(dft):
    for j,token in enumerate(dft[shn].iloc[0:,0]):
        if token not in pattern:
            data=[]
            tokens=token.split("-")
            for g in tokens:
                if g in boin:
                    score=list(dfr1.iloc[boin.index(g)])
                elif g in shiin:
                    score=list(dfr2.iloc[shiin.index(g)])
                elif g == "<s>":
                    score=[-1]*4
                elif g == ".":
                    score=[-2]*4
                else:
                    print("アウト")
                    print(shn)
                    score=[-3]*4
                data=data+score
            pattern.append(data) #toekn+パターンの調音
            goku.append(token)
#         print(token.split("-"))
print(pattern)
print(goku)
```

```python
with pd.ExcelWriter(wfilename) as writer:
    wdf=pd.DataFrame(pattern,index=goku, columns=["score"]*12)
    wdf.to_excel(writer,sheet_name="scoring")
```
