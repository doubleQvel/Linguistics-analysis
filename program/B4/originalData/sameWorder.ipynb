```python
filename="/Users/kazuki/Documents/Study/data/ryukyu/triFreq.xlsx"
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(filename,sheet_name=None)
print(df)
```

```python
sheetNames=list(df.keys())
print(sheetNames)
```

```python
print(df[sheetNames[0]]["語句"])
```

```python
#その地点の語句がどの程度各地点に存在するのか?
located={}
for data  in df[sheetNames[0]]["語句"]:
    for i,shn in enumerate(sheetNames):
        if data in list(map(str,df[shn]["語句"])):
            print("Yes")
            if data not in located:
                located[data]=1
            else:
                located[data]+=1
```

```python
print(located)
```

```python
print(len(located))
```

```python
sort_values=sorted(located.items(), key=lambda x: x[1],reverse=True)
x=np.linspace(1,len(located),len(located))
y=np.zeros(len(located))
sort_keys=[k for k,v in sort_values]
i=0
for k,v in sort_values:
#     x.append(k)
    y[i]=v
    i+=1
x1=x[0:20]
y1=y[0:20]
plt.figure(figsize=(20,10))
plt.rcParams["font.size"] = 18
plt.bar(x1,y1)
plt.xticks(x1,sort_keys[0:20])
# plt.bar(x,y)
# plt.xticks(x,sort_keys)
plt.tight_layout()
# plt.show()
pictname="a1.png"
plt.savefig(pictname)
plt.close()
```
