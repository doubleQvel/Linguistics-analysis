## <a id="subsection2" href="#subsection2">2.変化後から似ている地点はどれか探す</a>

<p class="ec__link-index"><a href="#section2">[↑ 小目次へ]</a></p>

* 単純にcos類似度で調べるだけなんだ

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#類似度を求める
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
#全地点
allhindovector=[]
for si,shN in enumerate(shNs):
    allhindovector.append([])
    a=triDf[shN]["頻度"].values
    for sj,(locate,shN2) in enumerate(zip(locates,shNs)):
        if si == sj:
            continue
        b=triDf[shN2]["頻度"].values
        allhindovector[si].append([locate,cos_sim(a,b)])
```

```python
for si,shN in enumerate(shNs):
    print("{}のcos類似度".format(locates[si]))
    for sj,shN2 in enumerate(shNs[1:]):
        print(allhindovector[si][sj])
```

```python
wfCos=basepath.fotmat("words/cossim.xlsx")
```

```python
with pd.ExcelWriter(wfCos) as writer:
    for si,shN in enumerate(shNs):
        newLocates=[locate for locate,cossim in allhindovector[si]]
        wdf=pd.DataFrame(allhindovector[si],index=list(range(len(newLocates))), columns=["地点","類似度"])
        wdf.to_excel(writer,sheet_name=shN)
```
