```python
for pi,pattern in enumerate(patterns):
    moto,hikaku=pattern.split("->")
    moto=moto.split(" ")
    hikaku=hikaku.split(" ")
    for cj,consonant in enumerate(consonantsList):
        kekka=[i for i,oto in enumerate(moto) if oto == consonant]
        for conindex in kekka:
            if moto[conindex] == hikaku[conindex]:
                consonantsDic[cj][0]+=sum(tri3Df["hindo"].iloc[pi,0:].tolist())
            else:
                consonantsDic[cj][1]+=sum(tri3Df["hindo"].iloc[pi,0:].tolist())
```

```python
kiritume=[]
for ci,consonant in enumerate(consonantsList):
    if consonantsDic[ci][0] == 0 and  consonantsDic[ci][1]==0:
        kiritume.append(consonant)
```

```python
print(kiritume)
```

```python
consonantsloss=[consonant for consonant in consonantsList]
consonantsDiclost=[consonant for consonant in consonantsDic]
```

```python
for kiru in kiritume:
    for ci,consonant in enumerate(consonantsloss):
        if consonant == kiru:
            consonantsloss.pop(ci)
            consonantsDiclost.pop(ci)
```
