1地域をターゲットとして、他の地域を要素として、回数を見る。



```python
Tlist=[0 for i,w in enumerate(words)]
for i,word in enumerate(words):
    Tlist[i]=(~Orgdf[word]["1子音"].isin(["-9"])).to_list()
```

```python
Tdf=pd.DataFrame(Tlist,index=words,columns=locates)
```

```python
Tdf
```

```python
Clist=[[0 for j in locates]for i in words]
for i,locate in enumerate(locates):
    for j,word in enumerate(words):
        if Tdf.loc[word,locate]==True:
#             print(Tdf.loc[word,:].sum())
            Clist[j][i]=Tdf.loc[word,:].sum()
        else:
            Clist[j][i]=0
```

```python
Cdf=pd.DataFrame(Clist,index=words,columns=locates)
```

```python
Cdf
```

```python
for i in locates:
    print((~Cdf[i].isin([0])).sum())
```

```python
Klist=[[0 for j in locates]for i in locates]
for i,locate in enumerate(locates):
    for j,word in enumerate(words):
        if Tdf.loc[word,locate]==False:
            continue
        for k,locate2 in enumerate(locates):
            if locate==locate2:
                continue
            if Tdf.loc[word,locate2]==True:
                Klist[i][k]+=1
```
