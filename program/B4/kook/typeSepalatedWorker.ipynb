### トライグラムの生成

```python
#トライグラム
#単語からのでてくる全パターンを生成
def ngrams(text, n):
    #text:子音と母音の塊
    #[i:i+n]スライス の前にtext最初に<s>を追加
    #list.index(".")で最初の位置を確認　その位置まではスライスを行う if文で？
    text_list = ["<s>"]+list(text)
    if "." not in text_list:
        text_list.append(".")
    saku=text_list.index(".")
    return [text_list[i:i + n] for i, char in enumerate(text_list) if saku+1 >= i+n]
def ngramize(text, n):
    text_list = ["<s>"]+list(text)
    if "-1" in text_list: #-1は無くして圧縮を行う
        text_list=[i for i in text_list if i != "-1"]
    if "." not in text_list:
        text_list.append(".")
    saku=text_list.index(".")
    return [text_list[i:i + n] for i, char in enumerate(text_list) if saku+1 >= i+n]
```

### 全語彙に対応

```python
results=[]
allchanger=[]
for pi,page in enumerate(pages):
    wf=basepath.format("words/"+page+".xlsx")
    goiDf=pd.read_excel(wf,sheet_name=None,header=0, index_col=0)
    for ti,(locate,titen) in enumerate(zip(locates[1:],titens[1:])):
        if pi == 0:
            results.append([])
        print(page,locate,len(goiDf[soutei]),len(goiDf[locate]))
        for i in range(len(goiDf[locate])):
            if "-9" in map(str,goiDf[locate].iloc[i,].tolist()):
                continue
            moto=" ".join(map(str,goiDf[soutei].iloc[i,].tolist()))
            hikaku=" ".join(map(str,goiDf[locate].iloc[i,].tolist()))
            results[ti].append(moto+"->"+hikaku)
            if moto+"->"+hikaku not in allchanger:
                allchanger.append(moto+"->"+hikaku)
```

```python
print(len(allchanger))
```

```python
counter=[]
for li,locate in enumerate(locates[1:]):
        counter.append([])
        for gokei in allchanger:
            count=results[li].count(gokei)
            counter[li].append([gokei,count])
```
