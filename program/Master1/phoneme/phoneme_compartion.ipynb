```python
voweldf.drop(index=sakuzyokey[40:],inplace=True)
```

```python
voweldf
```

```python
vwl_com=[[0 for i in range(len(voweldf.index))] for j in range(len(voweldf.index))]
```

```python
for i in range(len(voweldf.index)):
    a=list(voweldf.iloc[i,1:])
    for j in range(i,len(voweldf.index)):
        b=list(voweldf.iloc[j,1:])
        c=0
        for a1,b1 in zip(a,b):
            if a1!=b1:
                c+=1
#         print(a,b,c)
        vwl_com[j][i]=c
```

```python
vwlcomDf=pd.DataFrame(vwl_com,index=list(voweldf.iloc[0:,0]),columns=list(voweldf.iloc[0:,0]))
```
