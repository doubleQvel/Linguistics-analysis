配列の構成設定

```python
vowels=[0 for i in range(0,7+1)]+[0 for i in range(0,5+1)]+[0 for i in range(0,4+1)]+[0 for i in range(0,3+1)]+[0 for i in range(0,2+1)]
consonants=[0 for i in range(0,14+1)]+[0 for i in range(0,7+1)]+[0 for i in range(0,5+1)]+[0 for i in range(0,2+1)]+[0 for i in range(0,1+1)]
```

```python
ncs=[14,7,5,2,1]
nvs=[7 ,5,4,2,2]
avcs=[14,7,7,5,5,4,2,2,1,2]
```

```python
print(sum(ncs),sum(nvs))
print(sum(avcs))
```

```python
#地域ごとの特徴ベクトル作成
fvs=[[0 for i in range(sum(nvs)+len(nvs)+sum(ncs)+len(ncs)) ] for il, loc in enumerate(locations) ]
for fn in filenames[20:40]:
    fnp=basepath.format("CSVtest20190311/"+fn)
    print(fnp)
    testdf=pd.read_csv(fnp,header=0,index_col=0,usecols=[0]+list(range(27,27+(10*8))),encoding="shift-jis")
    for il,loc in enumerate(locations):
        scores=list(testdf.iloc[il+2,:])
        scores=[s.replace(".","-2") for s in scores]
        scores=list(map(int,scores))
#         print(scores)
#         print(len(scores))
        exscore=0
        for nscore,score in enumerate(scores):

            if nscore%10 == 0:
                exscore=0
            else:
                exscore+=avcs[nscore%10-1]
#             print(nscore,exscore,score)
            if score== -9 or score == -1 or score == -2 :
                continue;
            fvs[il][score+exscore]+=1
```

```python
fvsdf=pd.DataFrame(fvs)
```

```python
color_set=["#000000","#800000",
    "#ff0000","#800080",
    "#ff8c00","#ff00ff",
    "#008000","#808000",
    "#000080","#ff6633"]
```

```python
himethod='ward'
newdata=fvsdf
# newdata=newdata.T
pdis=dist.pdist(newdata,'cosine')
z = hierarchy.linkage(pdis, method=himethod,metric="cosine")
```

```python
#図の生成
figtitle='音素情報による地域のクラスタリング'
# savefile=basepath.format("gram3/pic/dendAE.png")

plt.figure(figsize=(60,25))
plt.rcParams['font.family'] = 'IPAexGothic'
plt.rcParams['font.size'] = 50
plt.title(figtitle, fontsize=60)
plt.xlabel('Regions', fontsize=55)
plt.ylabel('Intercluster distance', fontsize=55)
hierarchy.set_link_color_palette(color_set) # ６クラスタまでの色を指定
hierarchy.dendrogram(z, leaf_font_size=35., # 横軸の文字の大きさを指定
                        color_threshold=0.3 * np.max(z[:, 2]), # ユークリッド平方距離が７以上を同色で表示
                        above_threshold_color='black',
                        labels=tuple(locations)) # ユークリッド平方距離が７以上を黒色に指定
plt.tight_layout()
# plt.savefig(savefile)
plt.show()
plt.close()
```

```python
threshold = 0.3 * np.max(z[:, 2])
c = hierarchy.fcluster(z, threshold, criterion='distance')
print(c)
# wf=basepath.format("gram3/counter/AE.csv")
wdf=pd.DataFrame(c, columns=["市町村"])
# wdf["グループ"]=c
# wdf.to_csv(wf,header=False, index=False) #シート名[地点名]
```

```python
km = KMeans(n_clusters=3,            # クラスターの個数
            init='k-means++',           # セントロイドの初期値をランダムに設定  default: 'k-means++'
            n_init=20,               # 異なるセントロイドの初期値を用いたk-meansの実行回数 default: '10' 実行したうちもっとSSE値が小さいモデルを最終モデルとして選択
            max_iter=300,            # k-meansアルゴリズムの内部の最大イテレーション回数  default: '300'
            tol=1e-04,               # 収束と判定するための相対的な許容誤差 default: '1e-04'
            random_state=0)          # セントロイドの初期化に用いる乱数発生器の状態

y_km = km.fit_predict(fvs)
```

```python
print(y_km)
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib import cm

cluster_labels = np.unique(y_km)       # y_kmの要素の中で重複を無くす
n_clusters=cluster_labels.shape[0]     # 配列の長さを返す。つまりここでは n_clustersで指定した3となる

# シルエット係数を計算
silhouette_vals = silhouette_samples(fvs,y_km,metric='cosine')  # サンプルデータ, クラスター番号、ユークリッド距離でシルエット係数計算
y_ax_lower, y_ax_upper= 0,0
yticks = []

for i,c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km==c]      # cluster_labelsには 0,1,2が入っている（enumerateなのでiにも0,1,2が入ってる（たまたま））
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)              # サンプルの個数をクラスターごとに足し上げてy軸の最大値を決定
        color = cm.jet(float(i)/n_clusters)               # 色の値を作る
        plt.barh(range(y_ax_lower,y_ax_upper),            # 水平の棒グラフのを描画（底辺の範囲を指定）
                         c_silhouette_vals,               # 棒の幅（1サンプルを表す）
                         height=1.0,                      # 棒の高さ
                         edgecolor='none',                # 棒の端の色
                         color=color)                     # 棒の色
        yticks.append((y_ax_lower+y_ax_upper)/2)          # クラスタラベルの表示位置を追加
        y_ax_lower += len(c_silhouette_vals)              # 底辺の値に棒の幅を追加

silhouette_avg = np.mean(silhouette_vals)                 # シルエット係数の平均値
plt.axvline(silhouette_avg,color="red",linestyle="--")    # 係数の平均値に破線を引く 
plt.yticks(yticks,cluster_labels + 1)                     # クラスタレベルを表示
plt.ylabel('Cluster')
plt.xlabel('silhouette coefficient')
plt.show()
```

```python
for ik in range(3,12):
    km = KMeans(n_clusters=ik,            # クラスターの個数
            init='k-means++',           # セントロイドの初期値をランダムに設定  default: 'k-means++'
            n_init=20,               # 異なるセントロイドの初期値を用いたk-meansの実行回数 default: '10' 実行したうちもっとSSE値が小さいモデルを最終モデルとして選択
            max_iter=300,            # k-meansアルゴリズムの内部の最大イテレーション回数  default: '300'
            tol=1e-04,               # 収束と判定するための相対的な許容誤差 default: '1e-04'
            random_state=0)          # セントロイドの初期化に用いる乱数発生器の状態
    y_km = km.fit_predict(fvs)
    
    cluster_labels = np.unique(y_km)       # y_kmの要素の中で重複を無くす
    n_clusters=cluster_labels.shape[0]     # 配列の長さを返す。つまりここでは n_clustersで指定した3となる

# シルエット係数を計算
    silhouette_vals = silhouette_samples(fvs,y_km,metric='cosine')  # サンプルデータ, クラスター番号、ユークリッド距離でシルエット係数計算
    y_ax_lower, y_ax_upper= 0,0
    yticks = []

    for i,c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km==c]      # cluster_labelsには 0,1,2が入っている（enumerateなのでiにも0,1,2が入ってる（たまたま））
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)              # サンプルの個数をクラスターごとに足し上げてy軸の最大値を決定
        color = cm.jet(float(i)/n_clusters)               # 色の値を作る
        plt.barh(range(y_ax_lower,y_ax_upper),            # 水平の棒グラフのを描画（底辺の範囲を指定）
                         c_silhouette_vals,               # 棒の幅（1サンプルを表す）
                         height=1.0,                      # 棒の高さ
                         edgecolor='none',                # 棒の端の色
                         color=color)                     # 棒の色
        yticks.append((y_ax_lower+y_ax_upper)/2)          # クラスタラベルの表示位置を追加
        y_ax_lower += len(c_silhouette_vals)              # 底辺の値に棒の幅を追加

    silhouette_avg = np.mean(silhouette_vals)                 # シルエット係数の平均値
    plt.axvline(silhouette_avg,color="red",linestyle="--")    # 係数の平均値に破線を引く 
    plt.yticks(yticks,cluster_labels + 1)                     # クラスタレベルを表示
    plt.ylabel('Cluster')
    plt.xlabel('silhouette coefficient')
    plt.savefig(basepath.format("/pic/cluster{}.png".format(str(ik))))
    plt.show()
    plt.close()
```
